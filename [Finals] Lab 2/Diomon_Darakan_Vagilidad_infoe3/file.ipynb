```python
import csv
import random
from faker import Faker

# Initialize the Faker library to generate fake English data
fake = Faker('en_US') 

# --- PREP: Generate Data ---
# Generate a list called raw_data containing 30 entries (not less than 25)
raw_data = []
for _ in range(30):
    entry = {
        'name': fake.name(), # Generates realistic English names
        'age': random.randint(22, 75),
        'zip_code': fake.zipcode()[:5], # Using standard US zipcode format
        'salary': random.randint(40000, 160000),
        'medical_risk': random.choice(["Low", "Medium", "High"])
    }
    raw_data.append(entry)

# --- LAB TASK ---

# 1. Isolating target & Calculate "Pre-Swap" Metrics
original_salaries = [entry['salary'] for entry in raw_data]

pre_total = sum(original_salaries)
pre_avg = pre_total / len(original_salaries)
print("--- PRE-SWAP METRICS ---")
print(f"Total Payroll: ${pre_total}")
print(f"Average Salary: ${pre_avg:.2f}\n")

# 2. The "Same-Value" Check & Random Shuffle
original_first_salary = raw_data[0]['salary']
shuffled_salaries = [entry['salary'] for entry in raw_data]

# Loop checks if the first person's new salary is equal to their old one, and re-shuffles if it is
while True:
    random.shuffle(shuffled_salaries)
    if shuffled_salaries[0] != original_first_salary:
        break

# 3. Re-injecting the Data
for i in range(len(raw_data)):
    raw_data[i]['salary'] = shuffled_salaries[i]

# 4. Verify Integrity
new_salaries = [entry['salary'] for entry in raw_data]
post_total = sum(new_salaries)
post_avg = post_total / len(new_salaries)

print("--- POST-SWAP METRICS (Verification) ---")
print(f"Total Payroll: ${post_total}")
print(f"Average Salary: ${post_avg:.2f}\n")

# 5. Export
with open('new.csv', mode='w', newline='', encoding='utf-8') as file:
    fieldnames = ['name', 'age', 'zip_code', 'salary', 'medical_risk']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    
    writer.writeheader()
    writer.writerows(raw_data)

print("Success: Anonymized data has been exported to 'new.csv'")
```
