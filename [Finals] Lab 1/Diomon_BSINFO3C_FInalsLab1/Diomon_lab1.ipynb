```python
def mask_pii(input_string, mask_char="X"):
    """
    Masks all sections of a hyphen-separated string except the last one.
    """
    print(f"\n--- Processing: {input_string} ---")
    
    # Step 1: Breakdown
    # Create a variable and use the split() method on the hyphen (-)
    parts = input_string.split("-")
    print(f"Step 1 (Breakdown): {parts}")
    
    # Step 2: Loop
    # Iterate through the list. We use len(parts) - 1 to ignore the last index (-1)
    for i in range(len(parts) - 1):
        # Replaces the digits with the mask_char while maintaining the original length
        parts[i] = mask_char * len(parts[i])
        
    print(f"Step 2 (Masked List): {parts}")
    
    # Step 3: Re-assembly
    # Use the "-".join() method to combine the list back into a single string
    masked_string = "-".join(parts)
    print(f"Step 3 (Re-assembly): {masked_string}")
    
    return masked_string

# --- Validation Tests ---

# Credit Card Test
cc_input = "4532-8812-9001-4321"
cc_result = mask_pii(cc_input)
print(f"Final Output: {cc_result}")

# Phone Number Test
phone_input = "123-456-7890"
phone_result = mask_pii(phone_input)
print(f"Final Output: {phone_result}")
```
