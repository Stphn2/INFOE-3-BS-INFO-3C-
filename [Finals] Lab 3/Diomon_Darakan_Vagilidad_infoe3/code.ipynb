```python
import csv
import copy

# 1. Read the raw data from the CSV generated in your previous lab
raw_data = []
with open('new.csv', mode='r', encoding='utf-8-sig') as file:
    reader = csv.DictReader(file)
    for row in reader:
        raw_data.append(dict(row))

# 2. Protect original data
new_data = copy.deepcopy(raw_data)
hit_list = ["name", "medical_risk"]

# 3. Execute suppression
for record in new_data:
    for key in hit_list:
        record.pop(key, None)

print("Verification First Record:", new_data[0])

# 4. Export to new CSV
output_filename = "privacy_cleaned_dataset.csv"
fieldnames = ["age", "zip_code", "salary"]

with open(output_filename, mode='w', newline='') as file:
    # extrasaction='ignore' is a safe way to ensure ONLY our 3 columns make it to the final file
    writer = csv.DictWriter(file, fieldnames=fieldnames, extrasaction='ignore')
    writer.writeheader()
    
    for record in new_data:
        writer.writerow(record)
        
print("Success: Exported to privacy_cleaned_dataset.csv")
```
